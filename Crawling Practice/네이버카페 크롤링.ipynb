```python
url_list = []

# 몇 페이지까지 글을 크롤링할지
num_page = 5
for i in range(1, num_page):  # i는 페이지 번호
    # query(검색어)가 입력된 url
    url = "https://cafe.naver.com/firenze?iframe_url=/ArticleSearchList.nhn%3Fsearch.clubid=10209062%26search.media=0%26search.searchdate=all%26search.defaultValue=1%26userDisplay=15%26search.option=0%26search.sortBy=date%26search.searchBy=1%26search.query=%C0%AF%B7%B4%26search.viewtype=title%26search.page={}".format(i)
    driver.get(url)
    time.sleep(0.5)
    
    # iframe 접근
    driver.switch_to_frame('cafe_main')
    
    # URL 크롤링 시작
    overlays = """.article"""
    article_raw = driver.find_elements_by_css_selector(overlays)               # category
    article_raw
    
    for article in article_raw:
        url = article.get_attribute('href')   
        url_list.append(url)
    
    print(i)

url_list
```

# 글 크롤링

```python
# "url_list.csv" 불러오기
url_load = pd.read_csv('url_list.csv')        # 기본 모델
url_load
```

```python
# 크롤링
dict = {}  # 전체 크롤링 데이터를 담을 그릇
```

```python
for i in tqdm(range(0, 15)):
    try:
        # 글 띄우기
        url = url_load['url'][i]
        driver.get(url)   # 최신글 띄우기
        time.sleep(2)
        
        # iframe 접근
        driver.switch_to_frame('cafe_main')

        target_info = {}   # 글 하나의 딕셔너리
    #     # 카테고리 크롤링 시작
    #     overlays = "#main-area .inbox .fl a.m-tcol-c"
    #     cate = driver.find_element_by_css_selector(overlays)      # category
    #     category = cate.text
    #     category

        # 제목 크롤링 시작
        overlays = ".title_text"                                 
        tit = driver.find_element_by_css_selector(overlays)         # title
        title = tit.text
        title
        
        # 내용(content) 크롤링 시작
        overlays = ".se-module.se-module-text"                                  
        cont = driver.find_element_by_css_selector(overlays)        # content
        content = cont.text
        content

        # 작성자(id) 크롤링 시작
        overlays = ".nickname"                                  
        id_ = driver.find_element_by_css_selector(overlays)         # person
        person = id_.text
        person

        # 날짜(date) 크롤링 시작
        overlays = ".date"                
        date_ = driver.find_element_by_css_selector(overlays)        # datetime
        datetime = date_.text
        datetime

        # 댓글(reply) 크롤링 시작
        overlays = ".text_comment"                                  
        reply = driver.find_elements_by_css_selector(overlays)      # re_list
        r_num = len(reply)
        re_list = []       
        for re in reply:    
            re_list.append(re.text)
        replies = ",".join(re_list)
        replies
        
#         # 이미지(image) 크롤링 시작
#         overlays = "article_img.ATTACH_IMAGE"                                 
#         img_tmp = driver.find_elements_by_name(overlays)         # image

#         img_list = []   
#         for img in img_tmp:    
#             img_url = img.get_attribute('src')
#             img_list.append(img_url)
#         image = ",".join(img_list)
#         image 

        # 글 1개 딕셔너리에 크롤링한 데이터 집어넣기
        #target_info['category'] = category
        target_info['title'] = title
        target_info['url'] = url
        target_info['content'] = content
        target_info['person'] = person
        target_info['datetime'] = datetime
        target_info['reply'] = replies
#         target_info['image'] = image

        # 글 전체 딕셔너리에 글 각각 딕셔너리 넣기
        # target_info

        dict[i] = target_info
        time.sleep(2)

        print(i, title)
    
    except:      # 에러나면 다음 글로 이동
        continue
```

```python
# 판다스로 만들기
import pandas as pd
result_df = pd.DataFrame.from_dict(dict, 'index')
result_df
```
