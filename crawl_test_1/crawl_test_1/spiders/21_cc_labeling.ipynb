```
import pandas as pd
import numpy as np
from IPython.display import clear_output
```

```
df = pd.read_csv('/content/gdrive/My Drive/프로젝트/cc_post_final/cc_id_region2.csv')
```

```
def label(df, start_idx=0):
    for x in range(start_idx, len(df)):
        print('현재 지역 라벨: ', df.loc[x][0])
        print([location for location in df.loc[x] if type(location)!=float][1:])
        region_label = input('사용자의 지역을 숫자로 입력해 주세요(종료를 원하시면 0을 입력해 주세요)\n수도권(1), 강원(2), 충청(3), 전라(4), 경상(5), 모름(6) ')
        if region_label == '0':
            print('종료 인덱스: ', x)
            break
        df['region'][x] = region_label
        clear_output()
    print('더 이상 라벨링할 데이터가 없습니다!')
```

```
# (데이터프레임, 라벨링 시작하고 싶은 인덱스)를 함수에 넣어서 실행
label(df, 4)
```
