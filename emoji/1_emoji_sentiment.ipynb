```
import pandas as pd
import numpy as np
# from pandas import DataFrame as df
```

```
import json

contents_ls = []

for i in range(1, 144) : 
    with open("/content/gdrive/My Drive/프로젝트/emoji_json_file/{}.json".format(i), "r") as fp:
        temp_json =json.load(fp)
        print('{}/143'.format(i))
        for source_idx in range(len(temp_json)) :
            try :
                contents_ls.append(temp_json[source_idx]['content'])
            except :
                continue

contents_ls
```

```
import emoji

emoji_ls = []

for content in contents_ls :
    try :
        for char in content :
            if char in emoji.UNICODE_EMOJI :
                emoji_ls.append(char)       
    except :
        continue

emoji_ls
```

```
from collections import Counter

counter = Counter(emoji_ls)
print(counter)
```

```
counter['🏻'] # 업데이트 안된 이모티콘이 처리된모양으로 보임. 가장 많이 쓰인 이모티콘이 아니므로 무시 가능
```

```
emoji_top100 = counter.most_common(n=100) #자주 쓰인 이모티콘 Top 100위
emoji_top100
```

```
emoji_df = pd.DataFrame(emoji_top100, columns = ['emoji', 'e_count'], index = range(1, 101))
emoji_df
```

```
emoji_df = emoji_df[['e_count','emoji']]
emoji_df
```

```
emoji_df['sentiment'] = np.nan
emoji_df['unicode'] = np.nan
emoji_df
```
